```python
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
from ultralytics import YOLO
import cv2
import numpy as np

model = YOLO("yolo11n.pt")
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()
    flipped_frame = cv2.flip(frame, 1)

    if not ret:
        break

    results = model.predict(source=flipped_frame, conf=0.5)

    annotated_frame = results[0].plot()
    cv2.imshow("video", annotated_frame)

    if cv2.waitKey(1) != -1:
        break

cap.release()
cv2.destroyAllWindows()
```
