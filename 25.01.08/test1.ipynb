# 실습 1 캠화면에 적용


## 1. 캠화면에서 얼굴과 눈을 찾아서 얼굴 및 눈 영역에 사각형 표시


```python
import cv2

face_cascade = cv2.CascadeClassifier(
    "../cascade/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("../cascade/haarcascade_eye.xml")
```

```python
import cv2

face_cascade = cv2.CascadeClassifier(
    "../cascade/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("../cascade/haarcascade_eye.xml")

cap = cv2.VideoCapture(0)

fps = cap.get(cv2.CAP_PROP_FPS)

if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()

    if not ret:
        break
    frame_flip = cv2.flip(frame, 1)
    gray = cv2.cvtColor(frame_flip, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    eyes = eye_cascade.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(20, 20))

    for x, y, w, h in faces:
        cv2.rectangle(frame_flip, (x, y), (x+w, y+h),
                      (12, 45, 86), 2, cv2.LINE_AA)
    for x, y, w, h in eyes:
        cv2.rectangle(frame_flip, (x, y), (x+w, y+h),
                      (125, 85, 100), 2, cv2.LINE_AA)

    cv2.imshow("detect", frame_flip)

    if cv2.waitKey(1) != -1:
        break

cap.release()
cv2.destroyAllWindows()
```

## 2. 네모 위에 "Face", "Eye"라는 글자 띄우기


```python
import cv2

face_cascade = cv2.CascadeClassifier(
    "../cascade/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("../cascade/haarcascade_eye.xml")

cap = cv2.VideoCapture(0)

fps = cap.get(cv2.CAP_PROP_FPS)

if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()

    if not ret:
        break
    frame_flip = cv2.flip(frame, 1)
    gray = cv2.cvtColor(frame_flip, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(20, 20))

    eyes = eye_cascade.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), maxSize=(50, 50))

    for x, y, w, h in faces:

        cv2.rectangle(frame_flip, (x, y), (x+w, y+h),
                      (12, 45, 86), 2, cv2.LINE_AA)
        cv2.putText(frame_flip, "face", (x, y),
                    cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
    for x, y, w, h in eyes:
        cv2.rectangle(frame_flip, (x, y), (x+w, y+h),
                      (125, 85, 100), 2, cv2.LINE_AA)
        cv2.putText(frame_flip, "eye", (x, y),
                    cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)

    cv2.imshow("detect", frame_flip)

    if cv2.waitKey(int(1000/fps)) != -1:
        break

cap.release()
cv2.destroyAllWindows()
```
