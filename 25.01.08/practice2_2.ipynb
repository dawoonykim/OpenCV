```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("yolo11n.pt")
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break
    coppied = cv2.flip(frame, 1)

    results = model.predict(source=coppied, conf=0.5)
    annotated_frame = results[0].plot()

    cv2.imshow("camera", annotated_frame)

    if cv2.waitKey(1) != -1:
        break
cap.release()
cv2.destroyAllWindows()
```
