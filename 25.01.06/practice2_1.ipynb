```python
import cv2
import numpy as np
# 스캔에 사용할 이미지를 입력합니다
img = cv2.imread("../images/business card.jpg")
# 포인트를 저장할 리스트를 생성합니다
resize_img = cv2.resize(img, None, fx=0.45, fy=0.45)
point_list = []

# 최종 결과를 출력하는 함수입니다


def show_result():
    width, height = (point_list[1][0]-point_list[0]
                     [0]), (point_list[3][1]-point_list[0][1])

    src = np.float32(point_list)
    dst = np.float32([[0, 0], [width, 0], [width, height], [0, height]])

    mat = cv2.getPerspectiveTransform(src, dst)

    result = cv2.warpPerspective(resize_img, mat, (width, height))

    cv2.imshow("result", result)


flag = False
# 마우스 이벤트와 함께 실행되는 콜백함수입니다.


def onMouse(event, x, y, flags, _):
    # 조건을 만족하면 결과를 출력하는 함수를 실행합니다.
    global flag
    copied = resize_img.copy()
    if event == cv2.EVENT_LBUTTONDOWN and len(point_list) < 4:
        flag = True
        point_list.append([x, y])

    if flag == True:
        prev_point = None

        for point in point_list:
            cv2.circle(copied, point, 5, (0, 0, 0), 3, cv2.LINE_AA)

            if prev_point:
                cv2.line(copied, prev_point, point, (0, 0, 0), 3, cv2.LINE_AA)

            prev_point = point
        next_point = (x, y)
        if len(point_list) == 4:
            next_point = point_list[0]
            show_result()
        cv2.line(copied, prev_point, next_point, (0, 0, 0), 3, cv2.LINE_AA)
        cv2.imshow(name, copied)


# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, onMouse)
cv2.imshow(name, resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
