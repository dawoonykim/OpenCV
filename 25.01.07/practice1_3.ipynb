```python
img = cv2.imread(card)

name = "Card"
cv2.namedWindow(name)
coppied = img.copy()
gray = cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)

ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

contours, hierachy = cv2.findContours(
    binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    x, y, width, height = cv2.boundingRect(contour)

    cv2.rectangle(coppied, (x, y), (x+width, y+height),
                  (45, 145, 122), 2, cv2.LINE_AA)

cv2.imshow(name, coppied)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread(card)

name = "Card"
cv2.namedWindow(name)
coppied = img.copy()
gray = cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)

ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

contours, hierachy = cv2.findContours(
    binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

contour_list = []
for contour in contours:
    # print(cv2.contourArea(contour))
    if cv2.contourArea(contour) > 2000:
        contour_list.append(contour)
count=1
for contour in contour_list:
  x, y, width, height = cv2.boundingRect(contour)
  cv2.rectangle(coppied, (x, y), (x+width, y+height),
                  (45, 145, 122), 2, cv2.LINE_AA)
  cv2.imshow(name, coppied)
  cropped = img[y:y+height, x:x+width]

  cv2.imshow(f"cropped_{count}", cropped)
  count+=1
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 같은 창에서 표시

```python
img = cv2.imread(card)

name = "Card"
cv2.namedWindow(name)
coppied = img.copy()
gray = cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)

ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

contours, hierachy = cv2.findContours(
    binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

contour_list = []
for contour in contours:
    # print(cv2.contourArea(contour))
    if cv2.contourArea(contour) > 2000:
        contour_list.append(contour)

cv2.createTrackbar("index", name, 0, len(contour_list)-1, lambda x: x)

while True:
    coppied = img.copy()
    index = cv2.getTrackbarPos("index", name)

    x, y, width, height = cv2.boundingRect(contour_list[index])
    cv2.rectangle(coppied, (x, y), (x+width, y+height),
                  (45, 145, 122), 2, cv2.LINE_AA)
    cv2.imshow(name, coppied)
    cropped = img[y:y+height, x:x+width]

    cv2.imshow("cropped", cropped)
    if cv2.waitKey(1) != -1:
        break
cv2.destroyAllWindows()
```
