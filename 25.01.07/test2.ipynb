```python
import cv2

img = cv2.imread("../images/playing_cards.png")

coppied = img.copy()

name = "playing_card"
cv2.namedWindow(name)

gray = cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours, hierachy = cv2.findContours(
    binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

filtered_contours = []
MIN_AREA = 20000
for contour in contours:
    # print(cv2.contourArea(contour))
    if cv2.contourArea(contour) > MIN_AREA:
        filtered_contours.append(contour)

cv2.createTrackbar("index", name, 0, len(filtered_contours)-1, lambda x: x)
while True:
    coppied = img.copy()
    index = cv2.getTrackbarPos("index", name)
    x, y, width, height = cv2.boundingRect(filtered_contours[index])
    cv2.rectangle(coppied, (x, y), (x+width, y+height),
                  (124, 62, 148), 2, cv2.LINE_AA)
    cv2.imshow(name, coppied)

    # 각 경계선마다의 좌표를 리스트에 저장
    # 저장 받은 좌표로 이미지 자르기
    coordinate_list = []
    for contour in contours:
        if cv2.contourArea(contour) > MIN_AREA:
            x, y, width, height = cv2.boundingRect(contour)
            # cv2.rectangle(coppied, (x, y), (x+width, y+height),
            # (0, 0, 255), 2, cv2.LINE_AA)
            coordinate_list.append([x, y, x+width, y+height])
    img_list = []
    for coordinate in coordinate_list:
        cropped = img[coordinate[1]:coordinate[3], coordinate[0]:coordinate[2]]
        img_list.append(cropped)

    for index in range(len(img_list)):
        cv2.imshow(f"cropped_{index+1}", img_list[index])

    if cv2.waitKey(1) != -1:
        break
print(coordinate_list)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("../images/many_cards.png")

coppied = img.copy()

name = "many playing_card"
cv2.namedWindow(name)

gray = cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours, hierachy = cv2.findContours(
    binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

filtered_contours = []
MIN_AREA = 2000
for contour in contours:
    # print(cv2.contourArea(contour))
    if cv2.contourArea(contour) > MIN_AREA:
        filtered_contours.append(contour)

cv2.createTrackbar("index", name, 0, len(filtered_contours)-1, lambda x: x)
while True:
    coppied = img.copy()
    index = cv2.getTrackbarPos("index", name)
    x, y, width, height = cv2.boundingRect(filtered_contours[index])
    cv2.rectangle(coppied, (x, y), (x+width, y+height),
                  (124, 62, 148), 2, cv2.LINE_AA)
    cv2.imshow(name, coppied)

    # 각 경계선마다의 좌표를 리스트에 저장
    # 저장 받은 좌표로 이미지 자르기

    x, y, width, height = cv2.boundingRect(filtered_contours[index])
    cropped = img[y:y+height, x:x+width]
    cropped_resize = cv2.resize(cropped, None, fx=1.5, fy=1.5)
    enhanced = cv2.detailEnhance(cropped_resize, sigma_s=10, sigma_r=0.15)
    cv2.imshow("cropped", enhanced)

    if cv2.waitKey(1) != -1:
        break

cv2.destroyAllWindows()
```
